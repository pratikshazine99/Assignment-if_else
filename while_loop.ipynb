```python
#Guess the Number:
#Write a guessing game where the user keeps guessing a predefined number until they get it right.
secret_number = 7  

guess = int(input("Guess the number: "))

while guess != secret_number:
    print("Wrong guess! Try again.")
    guess = int(input("Guess the number: "))

print("Congratulations! You guessed the correct number.")
```

```python
#Simple Sum Calculator:
#Keep asking the user for numbers and calculate the sum until they enter 0.
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    total += num

print("Sum =", total)
```

```python
#Intermediate Level
#Input a number and reverse its digits using a while loop.
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reverse number is:", reverse)
```

```python
#Find the Smallest Divisor:
#Write a program to find the smallest divisor of a given number other than 1
num = int(input("Enter a number: "))

divisor = 2

while num % divisor != 0:
    divisor = divisor + 1

print("Smallest divisor other than 1 is:", divisor)
```
